```python
# Build prompts

from langchain.chains import LLMChain
from langchain.prompts import (
    ChatPromptTemplate,
    HumanMessagePromptTemplate,
    MessagesPlaceholder,
    SystemMessagePromptTemplate,
)
from langchain.memory import ConversationSummaryBufferMemory
from langchain.memory import ConversationBufferMemory
from langchain.schema import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAI


# Create actual prompt using the prompt string
def create_chat_prompt(prompt_str):
  return ChatPromptTemplate(
      messages=[
        SystemMessagePromptTemplate.from_template(prompt_str),
        # The `variable_name` here is what must align with memory
        MessagesPlaceholder(variable_name="chat_history"),
        # Directly put the question in as the human message
        # Future usage: input question as a variable by {question: "My question?"}
        HumanMessagePromptTemplate.from_template("{question}"),
      ]
  )


prompt_1 = create_chat_prompt(prompt_str_1)
prompt_2 = create_chat_prompt(prompt_str_2)
prompt_3 = create_chat_prompt(prompt_str_3)
prompt_4 = create_chat_prompt(prompt_str_4)
prompt_5 = create_chat_prompt(prompt_str_5)
```

```python
# Evaluators for checking the response generated by llm.
# They grades answers to questions directly using an LLM

from langchain.evaluation import load_evaluator

qa_evaluator = load_evaluator("qa")

lc_evaluator_correctness = load_evaluator("labeled_criteria", criteria="correctness")
lc_evaluator_harmfulness = load_evaluator("labeled_criteria", criteria="harmfulness")
```

```python
from langchain.chains import ConversationChain
import pandas as pd

# For prompt evaluation
score_board = pd.DataFrame(columns=['model', 'prompt', 'question', 'answer', 'score'])

# Create llm
llm_1 = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
llm_2 = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.5)
llm_3 = ChatOpenAI(model="gpt-3.5-turbo", temperature=1)

gpt4_1 = ChatOpenAI(model="gpt-4", temperature=0)
gpt4_2 = ChatOpenAI(model="gpt-4", temperature=0.8)

models = {
  'GPT3.5-low_temp': llm_1,
  'GPT3.5-mid_temp': llm_2,
  'GPT3.5-hig_temp': llm_3,
  'GPT4-2-low_temp': gpt4_1,
  'GPT4-2-hig_temp': gpt4_2,
}
prompt_list = {
  "Bonnie girl": prompt_1,
  "Dragon boy": prompt_2,
  "Custom Instruction": prompt_3,
  # "Super Prompt": prompt_4,
  "LearnOS": prompt_5
}
```

```python
for model in models:
  for prompt in prompt_list:
    for question_round in question_list:
      memory_buffer = ConversationBufferMemory(memory_key="chat_history", return_messages=True)
      conversation = LLMChain(llm=models[model], prompt=prompt_list[prompt], verbose=True, memory = memory_buffer)
      output = ''
      for i in range(len(question_round)):
        if i != len(question_round)-1:
          output = conversation({"question": question_round[i]})['text'] # retain the last output
        ground_truth = question_round[i]

      print(output)

      eval_result = lc_evaluator_correctness.evaluate_strings(
        input= question_round,
        prediction= output,
        reference= ground_truth,
      )

      new_row = {'model': model, 'prompt': prompt, 'question': question_round, 'answer':output, 'score': eval_result["score"]}
      score_board = score_board.append(new_row, ignore_index=True)
```

Here I get the following conclusion:

1.   The prompt extraction attacks trigger some deep mechanisms of LLMs, making models give out the prompt contents before analyzing them. This makes it almost impossible to defend it within the prompt.
2.   Only an outside source can effectively prevent the prompt leaking.

Two methods of defense:

1.   Using another LLM to determine whether the prompt is leaked.

  Pros
  *   Very wise, almost block all attacks, and can maintain normal conversations.
  *   If there's a leak, the modifier model can craft another response that correctly answers the question.

  Cons

  *   Slow down the response, especially under stream processing. (This is sometimes not acceptable, because response time is the top concern of engineers)


2.   Inserting special characters into the prompt and stop generating when detecting the special characters.

  Pros

  *   Doesn't affect response speed.
  *   Can specify which part of the prompt is secret.

  Cons

  *   Might affect the performance of the prompt
  *   Sometimes generates very weird responses. (These two are not very severe, since they almost only happen when there's a real attack)
  *   Cannot 100% block attack, since the model does not always output prompt content as their original format. (This is the biggest weakness of this method)
